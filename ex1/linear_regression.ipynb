## Gradient Descent

<img src="data/img/cost_function.jpg" width="30%" height="10%" align="left"><img src="data/img/gradient_descent.jpg" width="60%" align="left">

```python
iterations = 1500
alpha = 0.01
theta = np.zeros([X.shape[1], 1])
```

```python
def h(theta, X):
    """假设函数h"""
    return np.dot(X, theta)
```

```python
def computeCost(X, y, theta):
    """代价函数J"""
    m = X.shape[0]
    h_y = h(theta, X) - y
    return (1./(2*m)) * np.sum(np.square(h_y))
```

```python
# testing of the cost function
J = computeCost(X, y, np.zeros((2, 1)))
print('With theta = [0 ; 0]; Cost computed = %f' % J)
print('Expected cost value (approx) 32.07\n')

J = computeCost(X, y, np.array([[-1], [2]]))
print('With theta = [-1 ; 2]; Cost computed = %f' % J)
print('Expected cost value (approx) 54.24\n')
```

```python
def gradientDescent(X, y, theta, alpha, iterations):
    """梯度下降算法"""
    J_history = np.zeros(iterations)
    m = X.shape[0]
    for iter in range(iterations):
        J_history[iter] = computeCost(X, y, theta)
        if iter<10:
            print('iter: %d, cost: %f, theta:' % (iter+1, J_history[iter]), theta.T)
        h_y = h(theta, X) - y
        theta = theta - alpha * (1./m) * np.transpose(np.dot(h_y.T, X))
    return theta, J_history
```

```python
theta, J_history = gradientDescent(X, y, theta, alpha, iterations)
```

```python
plt.subplot(211)
plt.plot(X[:,1], y, 'rx')
plt.plot(X[:,1], h(theta, X), '-', color='black')
plt.title('Scatter plot of training data')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')

plt.subplot(212)
plt.plot(J_history)
plt.title('cost function plot')
plt.xlabel('num of iterations')
plt.ylabel(r'$J(\theta)$')

# 自动调整子图参数和子图间隔
plt.tight_layout()
plt.show()
```

## Visualizing J(θ)

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
size = 100

theta0_vals = np.linspace(-10, 10, size)
theta1_vals = np.linspace(-1, 4, size)
J_vals = np.zeros([size, size])
for i in range(size):
    for j in range(size):
        theta_vals = np.array([[theta0_vals[i]], [theta1_vals[j]]]).reshape(-1,1)
        J_vals[j, i] = computeCost(X, y, theta_vals)

theta0_vals, theta1_vals = np.meshgrid(theta0_vals, theta1_vals)

ax.plot_surface(theta0_vals, theta1_vals, J_vals, rstride=1, cstride=1, cmap=plt.cm.coolwarm)
ax.set_xlabel(r'$\theta_0$', color='r')
ax.set_ylabel(r'$\theta_1$', color='g')
ax.set_zlabel(r'$J(\theta)$', color='b')
plt.title(r'$J(\theta)$ surface plot')
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_subplot(111)
ax.set_xlabel(r'$\theta_0$', color='r')
ax.set_ylabel(r'$\theta_1$', color='g')

C = ax.contour(theta0_vals, theta1_vals, J_vals, np.logspace(-2,3,20))
plt.clabel(C, inline=1, fontsize=10)

ax.plot(theta[0,0], theta[1,0], 'rx', markersize=10, linewidth=1)
plt.show()
```
